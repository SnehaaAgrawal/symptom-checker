```python
import pandas as pd
from sklearn.utils import shuffle

# Load the first dataset
df = pd.read_csv("/content/dataset.csv")

# Clean column names by replacing underscores with spaces and fill NaN values with 0
for col in df.columns:
    df[col] = df[col].str.replace('_', ' ')
df = df.fillna(0)

# Shuffle the DataFrame
df = shuffle(df)

# Drop duplicate rows
df = df.drop_duplicates()

# Convert DataFrame to a list of lists and remove 0 values
data_list = df.values.tolist()
data_list_cleaned = [[x for x in inner_list if x != 0] for inner_list in data_list]

# Load the descriptions dataset
descriptions_df = pd.read_csv("/content/symptom_Description.csv")

# Create a dictionary mapping diseases to descriptions
disease_description_map = dict(zip(descriptions_df["Disease"], descriptions_df["Description"]))

# Load the precautions dataset
precautions_df = pd.read_csv("/content/symptom_precaution.csv")

# Create a dictionary mapping disease to precautions
disease_precaution_map = {}
for index, row in precautions_df.iterrows():
    disease = row["Disease"]
    precautions = [row["Precaution_1"], row["Precaution_2"], row["Precaution_3"], row["Precaution_4"]]
    disease_precaution_map[disease] = precautions

# Define the symptoms
symptoms = ['high fever', 'headache']

# Initialize a list to store diseases matching the symptoms
diseases_matching_symptoms = []

# Iterate over the cleaned data list to find diseases matching the symptoms
for row in data_list_cleaned:
    if all(symptom in str(row) for symptom in symptoms):
        diseases_matching_symptoms.append(row)

# Initialize a set to store unique diseases
unique_diseases = set()

# Initialize a list to store unique disease descriptions and precautions
diseases_with_details = []

# Iterate over diseases matching the symptoms to extract unique diseases with descriptions and precautions
for disease in diseases_matching_symptoms:
    disease_name = disease[0]  # Assuming the disease name is the first element
    if disease_name not in unique_diseases:
        # Get the description and precautions for the current disease
        description = disease_description_map.get(disease_name, "Description not found")
        precautions = disease_precaution_map.get(disease_name, ["Precaution not found"] * 5)
        diseases_with_details.append(disease + [description] + precautions)
        unique_diseases.add(disease_name)

# Print the unique diseases with descriptions and precautions
print("Unique Diseases with Descriptions and Precautions:")
for disease in diseases_with_details:
    print(disease)

# Now you can use diseases_with_details as needed in your application.
```
