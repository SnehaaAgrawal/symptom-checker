```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from flask import Flask, jsonify
import random
```

```python
df = pd.read_csv(r"/content/dataset.csv")
for col in df.columns:
    df[col] = df[col].str.replace('_',' ')
df = df.fillna(0)
df = shuffle(df)#,random_state=42)
df.head()
```

```python
# Check for duplicate rows
print(len(df))
df = df.drop_duplicates()
print(len(df))
duplicate_rows = df[df.duplicated()]

# Print the duplicate rows
if not duplicate_rows.empty:
    print("Duplicate rows found:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

```python
a = df.values.tolist()
print(len(a))


b = [[x for x in inner_list if x != 0] for inner_list in a]
print(len(b))
# Now a_without_zeroes contains the lists from a with all zeroes removed
print(a[4])
print(b[4])
```

```python
print(len(b))
symptoms = [' cough']
diseases = []

for i in b:
  if all(j in str(i) for j in symptoms):
      diseases.append(i)

print(diseases)
print(len(diseases))

x = set()
dis = []
for j in diseases:
    f = j[0]  # Assuming the first attribute is in the first column
    if f not in x:
        dis.append(j)
        x.add(f)

print("Elements: ", len(dis), "\n")
print(dis)
result = dis
```

```python
#iterating the loop until only one entry exists, check if the next symptom on the list is true or not
    #we ask the user.
        #If they say yes, we do nothing
        #If they say no, we reduce the value by 0.5
        #If they say idk, we reduce the value by 0.75
        #If an entry has zero value, it is removed.

#display the disease along with other details

qn = ['Are you suffering from', 'Do you have', 'Are you feeling']
for i in dis:
  for j in i:
    if j in symptoms:
      i.remove(j)

def asx(sym):
  for i in dis:
    for j in i:
      if j in sym:
        i.remove(j)


while len(dis)>1:

  for i in dis:
    sym = random.choice(i[1:])
    choice = input(random.choice(qn) + sym  + " ")

    if choice == 'yes':
      for i in dis:
        if sym not in i:
          dis.remove(i)
          print("Removed", i[0])
          asx(sym)

    elif choice == 'no':
      for i in dis:
        if sym in i:
          dis.remove(i)
          print("Removed", i[0])
          asx(sym)


print("You are suffering from ", dis[0][0])
```
