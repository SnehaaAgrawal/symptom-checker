```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from flask import Flask, jsonify
import random
```

```python
def preprocess(df):
  for col in df.columns:
      df[col] = df[col].str.replace('_',' ')

  cols = df.columns
  data = df[cols].values.flatten()

  s = pd.Series(data)
  s = s.str.strip()
  s = s.values.reshape(df.shape)

  df = pd.DataFrame(s, columns=df.columns)
  df.head()

  df = df.fillna(0)
  df.head()
  return df
```

```python
# Check for duplicate rows
df = pd.read_csv(r"/content/dataset.csv")
df = preprocess(df)
df = df.drop_duplicates()
```

```python
a = df.values.tolist()
b = [[x for x in inner_list if x != 0] for inner_list in a]
```

```python
symptoms = ['stiff neck']
diseases = []

for i in b:
  if all(j in str(i) for j in symptoms):
      diseases.append(i)

x = set()
dis = []
for j in diseases:
    f = j[0]
    if f not in x:
        dis.append(j)
        x.add(f)
```

```python
df = shuffle(df)#,random_state=42)
df.head()
```

```python
qn = ['Are you suffering from ', 'Do you have ', 'Are you feeling ']

def asx(sym):
  for i in dis:
    for j in i:
      if j in sym:
        i.remove(j)

asx(symptoms)
while len(dis)>1:

  for i in dis:
    sym = random.choice(i[1:])
    choice = input(random.choice(qn) + sym  + " ")

    if choice == 'yes':
      for i in dis:
        if sym not in i:
          dis.remove(i)
          asx(sym)

    elif choice == 'no':
      for i in dis:
        if sym in i:
          dis.remove(i)
          asx(sym)
```

```python
print("You are suffering from", dis[0][0])

print("Symptoms:")
for symptom in dis[0][1:]:
    print("•", symptom.title())

sd = pd.read_csv('/content/symptom_Description.csv')
sd = preprocess(sd)

symptom_description = sd.loc[sd['Disease'] == dis[0][0], 'Description'].iloc[0]
print("Symptom Description:")
print(symptom_description)

pr = pd.read_csv('/content/symptom_precaution.csv')
pr = preprocess(pr)

precautions = pr.loc[pr['Disease'] == dis[0][0]].iloc[:, 1:].values.tolist()
if precautions:
    print("Precautions:")
    for precaution in precautions[0]:
        print("•", precaution.capitalize())
else:
    print("No precautions found for", dis[0][0])
```
